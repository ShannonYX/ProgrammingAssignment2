```
import numpy as np
import matplotlib.pyplot as plt
```

```
a=list()
b=list()
Eout=list()
x = np.linspace(-1,1,200)
#calculate f function
f=x**2
fig = plt.figure()

for i in range(0,300):
  x1=np.random.uniform(-1,1)
  x2=np.random.uniform(-1,1)
  ai=x1+x2
  bi=-x1*x2
  a.append(ai)
  b.append(bi)
  #calculate and plot g(x) functions (yellow lines in the picture)
  g=ai*x+bi
  plt.plot(x,g,'y')  
  Eouti = (g-f)**2
  Eout.append(Eouti)

abar = np.mean(a)
bbar = np.mean(b)
#calcualte and plot g_bar (blue line in the picture)
gbar = abar*x+bbar
plt.plot(x,gbar,'b', label='gbar')
#calculate E[E_out]
EEout=np.mean(Eout)
print('EEout=:',EEout)
#calculate bias
bias=np.mean((gbar-f)**2) 
print('bias=',bias)
#calculate var
asquare=[i**2 for i in a]
bsquare=[i**2 for i in b]
ab = [a[i]*b[i] for i in range(0,len(a))]
varx = np.mean(asquare)*(x**2) + 2*np.mean(ab)*x +np.mean(bsquare)
var = np.mean(varx)
print('var=',var)
#compare E[Eout] with bias + var
print('E[Eout]-bias-var=',EEout-bias-var)
#plot f(x) (green line in the picture)
plt.plot(x, f,'g',label='f(x)')
plt.legend(loc='lower center' )
plt.show()
```
